```python
#Sales_Data_Analysis_Mini_Project

#Task-1
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#Task-2 a.
df = pd.read_excel('Superstore.xls')

pd.set_option('display.max_columns',None)

#2 b.
df.head()
```

```python
#2 c.
df.columns
```

```python
#2 d.
df.describe()
```

```python
#2 e.
df.isnull().any()
```

```python
#Task 3 a.
df[['Order Date', 'Ship Date']] = df[['Order Date', 'Ship Date']].apply(pd.to_datetime)

df[['Order Date', 'Ship Date']].dtypes
```

```python
#3 b.
df['Price Per Unit'] = df['Sales']/df['Quantity']

df.head()
```

```python
#Task 4 a.
df['Year_Month'] = df['Order Date'].dt.to_period('M')

df.head()
```

```python
#4 b.

df1 = df.groupby('Year_Month')['Sales'].sum()

df1 = pd.DataFrame(df1.reset_index())
df1.rename(columns = {'Sales' : 'Monthly_Sales'}, inplace = True)
df1.head()
```

```python
#4 c.
df1.plot(x = 'Year_Month', y = 'Monthly_Sales', kind = 'bar', figsize = (15,10))

#4 d.
plt.xlabel('Year_Month', fontsize = 16)
plt.ylabel('Monthly Sales',fontsize = 16)
plt.title('Monthly Revenue Observations', fontsize = 16)
plt.legend(loc = 'upper left')
```

```python
#Task 5 a.
df['Year'] = df['Order Date'].dt.to_period('Y')

df.head()
```

```python
#5 b.
df2 = df.groupby('Year')['Sales'].sum()

df2 = pd.DataFrame(df2.reset_index())
df2.rename(columns = {'Sales' : 'Yearly_Sales'}, inplace = True)
df2
```

```python
#5 c.
df2.plot(x = 'Year', y = 'Yearly_Sales', kind = 'bar')

#5 d.
plt.xlabel('Year', fontsize = 16)
plt.ylabel('Yearly Sales',fontsize = 16)
plt.title('Yearly Revenue Observations', fontsize = 16)
plt.legend()
```

```python
#Task 6 a.
df1['Monthly_Growth'] = df1['Monthly_Sales'].pct_change()
df1.fillna(0, inplace = True)

df1.head()
```

```python
#6 b.
m1 = df1['Monthly_Growth'].max()

m1
```

```python
#6 c. 
df1.plot(x = 'Year_Month', y = 'Monthly_Growth', marker = 'o', figsize = (15,10))

#6 d.
plt.ylabel('Monthly Growth',fontsize = 16)
plt.title('Growth', fontsize = 16)
plt.legend()
```

```python
#Task 10 a.
mx = df.groupby('City')['Customer Name'].count().idxmax()
mn = df.groupby('City')['Customer Name'].count().idxmin()

print('City with\nMax customers : {}\nMin customers : {}'.format(mx,mn))
```
